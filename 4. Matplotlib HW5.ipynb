Лучше использовать **plt.subplots()** для данных целей. Если у вас блоки расположены в разных строках, тогда постараться не делать мелкую ширину у графиков, минимум ширина должна быть в половину ширины экрана

```python
# можете использовать следующие значения
# либо выбрать свои данные, а также функции (cos/sin и т д)
N = 100

x = np.arange(N)

y_1 = np.sin(x / 10)
y_2 = np.sin(x / 5)
```

```python
plt.plot(x, y_1, color = 'blue', alpha = 0.5, marker = "1")
plt.plot(x, y_2, color = "purple", alpha = 0.5, linestyle = "dashdot")

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Plot")
plt.grid(visible = True)
```

```python
fig = plt.figure()

axes1 = fig.add_subplot(121)
axes1.plot(x, y_1, color = "pink", alpha = 0.7, marker = "|")
axes2 = fig.add_subplot(122)
axes2.plot(x, y_2, color = "black", alpha = 0.7, marker = "D")

for ax in fig.axes:
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.grid(True)
    ax.set_title("Plot")
    
fig.tight_layout()
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(221)
axes1.plot(x, y_1, color = "pink", alpha = 0.7, marker = "|")
axes2 = fig.add_subplot(224)
axes2.plot(x, y_2, color = "black", alpha = 0.7, marker = "D")

for ax in fig.axes:
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.grid(True)
    ax.set_title("Plot")
    
fig.tight_layout()
```
