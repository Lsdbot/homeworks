7) Вычислить собственные векторы и собственные значения матрицы - X_matrix

- np.linalg.eig(X_matrix) - для поиска собственных значений и собственных векторов матрицы - X_matrix

- Возвращает w и v, где w - собственные значения матрицы, v - собственные векторы матрицы

- Может вернуть несколько значений и векторов, тогда, чтобы посмотреть на один из вариантов:

```python
X_matrix = np.array([[-1, -6], [2, 6]])

w,v = np.linalg.eig(X_matrix)
print(w)
print(v)
```
