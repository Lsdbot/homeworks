```python
import pandas as pd
```

```python
df = %sql Select * from employees
df = pd.DataFrame(df)
df[:5]
```

```python
dfM = df[df["gender"] == "M"]
dfM[:5]
```

```python
dfF = df[df["gender"] == "F"]
dfF[:5]
```

```python
unique_titles = df.title.unique()
unique_titles
```

```python
df_dict = {title:pd.DataFrame() for title in unique_titles}
df_dict
```

```python
for key in df_dict.keys():
    df_dict[key] = df[df["title"]==key]
```

```python
for key in df_dict.keys():
    print(df_dict[key][:5])
```

```python
from sqlalchemy import create_engine
```

```python
conn = create_engine('sqlite:///EmployeeSQL.db')
```

```python
for key in df_dict.keys():
    df_dict[key].to_sql(name = f'{key}', con = conn, if_exists = "replace")
```
